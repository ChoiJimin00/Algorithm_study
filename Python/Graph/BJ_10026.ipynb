```python
from collections import deque

dx = [-1,1,0,0]
dy = [0,0,-1,1]

### 정상인이 보는 matrix
def bfs1(x,y): 
    q = deque()
    q.append((x,y))
    
    visited[x][y] = 1
    status = matrix[x][y]
    
    while q:
        x,y = q.popleft()
    
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx<0 or nx>=N or ny<0 or ny>=N:
                continue
            if matrix[nx][ny] != status:
                continue
            if (matrix[nx][ny]==status) and (visited[nx][ny]==0):
                q.append((nx,ny))
                visited[nx][ny] = 1
             
### 적록색약이 보는 matrix   
def bfs2(x,y):
    q = deque()
    q.append((x,y))
    
    visited[x][y] = 0
    status = matrix[x][y]
    
    while q:
        x,y = q.popleft()
    
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx<0 or nx>=N or ny<0 or ny>=N:
                continue
            if matrix[nx][ny] != status:
                continue
            if (matrix[nx][ny]==status) and (visited[nx][ny]==1):
                q.append((nx,ny))
                visited[nx][ny] = 0
                

N = int(input())
matrix = []

for _ in range(N):
    matrix.append(list(map(str,input())))
    
visited = [[0]*N for _ in range(N)]

normal = 0
for i in range(N):
    for j in range(N):
        if visited[i][j] == 0:
            bfs1(i,j)
            normal += 1

## 적록색약용 구역 설정
for i in range(N):
    for j in range(N):
        if matrix[i][j] == 'R':
            matrix[i][j] = 'G'
            
abnormal = 0
for i in range(N):
    for j in range(N):
        if visited[i][j] == 1:
            bfs2(i,j)
            abnormal += 1

print(normal,abnormal)
```
