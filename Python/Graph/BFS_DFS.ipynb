```python
## BJ_1260
from collections import deque

N,M,V = map(int,input().split())
graph = [[False]*(N+1) for _ in range(N+1)]

for _ in range(M): 
    node1, node2 = map(int,input().split())
    graph[node1][node2] = graph[node2][node1] = True
    
visited = [0]*(N+1)

# DFS 구현 - 재귀함수 이용
def dfs(V):
    visited[V] = 1
    print(V, end=' ')
    for i in range(1, N+1):
        if (visited[i]==0) and (graph[V][i]==1):
            dfs(i)
            
# DFS 구현 - stack 이용          
def dfs2(V): 
    stack = [V]
    
    while stack:
        V = stack.pop()
        
        if visited[V] == 0:
            print(V, end=' ')
            visited[V] = 1
        
        for i in range(N, 0, -1): # 방문할 수 있는 정점이 여러 개 일때, 작은 수 부터 방문
            if (visited[i] == 0) and (graph[V][i]==1):
                stack.append(i)
    
    
## BFS 구현 - queue 이용
def bfs(V):
    q = deque()
    q.append(V)
    visited[V] = 0
    
    while q:
        V = q.popleft()
        print(V, end=' ')
        for i in range(1, N+1):
            if (visited[i]==1) and (graph[V][i]==1):
                q.append(i)
                visited[i]=0 
    
dfs(V)
print()
bfs(V)
```
